```python
def download_image_with_name(url,name):
    local_filename = 'images/' + name + '.jpeg'
    r = requests.get(url, stream=True)
    print(r)
    if r.status_code == 404:
        return ''
    with open(local_filename, 'wb') as f:
        shutil.copyfileobj(r.raw, f)

    return local_filename
```
